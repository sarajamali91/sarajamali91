```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
from pandas import read_csv
from matplotlib import pyplot

series = read_csv('daily-min-temperatures.csv', header=0, index_col=0, parse_dates=True, squeeze=True)

print(series.head())

series.plot(figsize = (15,8))
pyplot.show()
```

```python
# Time Series Plot with data represented with dot (.)
series.plot(figsize = (15,8), style='k.')
pyplot.show()
```

```python
# Plot with yearly break-down
from pandas import read_csv
from pandas import DataFrame
from pandas import Grouper
from matplotlib import pyplot

#series = read_csv('daily-min-temperatures.csv', header=0, index_col=0, parse_dates=True, squeeze=True)

groups = series.groupby(Grouper(freq='A'))

years = DataFrame()
for name, group in groups:
    years[name.year] = group.values

#pyplot.figure(figsize=(16,12))
years.plot(subplots=True, legend=True, figsize=(16,14))
pyplot.show()
```

```python
# Histogram and Density Plots
series.hist(figsize=(14,8))
pyplot.show()

series.plot(kind='kde', figsize=(14,8))
pyplot.show()
```

```python
#Box and Whisker Plots
groups = series.groupby(Grouper(freq='A'))

years = DataFrame()
for name, group in groups:
    years[name.year] = group.values
    
years.boxplot(figsize=(14,8))
pyplot.show()
```

```python
from pandas import concat

one_year = series['1982']

groups = one_year.groupby(Grouper(freq='M'))

months = concat([DataFrame(x[1].values) for x in groups], axis=1)
months = DataFrame(months)
months.columns = range(1,13)

months.boxplot(figsize=(14,8))
pyplot.show()
```
